# 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#lets first connect to web driver
driver= webdriver.Chrome('chromedriver.exe')
#first get the webpage
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
try:
    ranking_of_vid=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranking_of_vid:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('No details available')
except StaleElementReferenceException:
    Rank.append('No details available')
```

```python
Name=[]
try:
    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('No details available')
except StaleElementReferenceException:
    Name.append('No details available')
```

```python
Name
```

```python
Artist
try:
    artists=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artists:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('No details available')
except StaleElementReferenceException:
    Artist.append('No details available')
Artist
```

```python
try:
    date_list=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in date_list:
        Upload_Date.append(i.text)
except NoSuchElementException:
    Upload_Date.append('No details available')
except StaleElementReferenceException:
    Upload_Date.append('No details available')
```

```python
Upload_Date
```

```python
try:
    list1=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in list1:
        Views.append(i.text)
except NoSuchElementException:#handling no such element exception
    Views.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    Views.append('No details available')
Views
```

```python
df=pd.DataFrame({})
df['Rank']=Rank
df['Name']=Name
df['Artist']=Artist
df['Upload_date']=Upload_Date
df['Views']=Views
df
```

# Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details:
A) Rank
B) State
C) GSDP(18-19)
D) GSDP(17-18)
E) Share(2017)
F) GDP($ billion)
Note: - From statisticstimes home page you have to reach to economy page through code.

```python
#lets first connect to web driver
driver= webdriver.Chrome('chromedriver.exe')
#first get the webpage http://statisticstimes.com/
driver.get('http://statisticstimes.com/')
```

```python
#clicking on economy
driver.find_element_by_xpath("//div[@class='navbar']/div[2]/button").click()
```

```python
#clicking on india
ind_clk=driver.find_element_by_xpath("//div[@class='dropdown-content']/a[3]")
driver.get(ind_clk.get_attribute('href'))
```

```python
#clicking on gdp of indian states
statewise=driver.find_element_by_xpath("//ul[@style='list-style-type:none;margin-left:20px;']/li[1]/a[1]")
statewise.click()
```

```python
# creating empty lists for scraping data
Rank=[] 
State=[] 
GSDP_2019_to_2020=[] 
GSDP_2018_to_2019=[] 
Share_2018_to_2019=[] 
GDP_Dollar_billion_=[]
```

```python
#scraping rank
rank = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[1]')
for i in rank:
    Rank.append(i.text)
#scraping state
state = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[2]')
for i in state:
    State.append(i.text)
#scraping gdsp_19_20
gdsp_19_20 = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[3]')
for i in gdsp_19_20:
    GSDP_2019_to_2020.append(i.text)
#scraping gsdp(18-19)
gdsp_18_19 = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[4]')
for i in gdsp_18_19:
    GSDP_2018_to_2019.append(i.text)
#scraping share
share = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[5]')
for i in share:
    Share_2018_to_2019.append(i.text)
#scraping gdp($ billion)
gdp_dollar = driver.find_elements_by_xpath('//table[@id="table_id"]/tbody/tr/td[6]')
for i in gdp_dollar:
    GDP_Dollar_billion_.append(i.text)
```

```python

df=pd.DataFrame({})
df['Rank']=Rank 
df['State']=State
df['GSDP(19-20)']=GSDP_2019_to_2020 
df['GSDP(18-19)']=GSDP_2018_to_2019 
df['Share(18-19)']=Share_2018_to_2019 
df['GDP($ billion)']=GDP_Dollar_billion_
df
```

# Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.


```python
#lets first connect to web driver
driver= webdriver.Chrome('chromedriver.exe')
#first get the webpage  https://github.com/
driver.get('https://github.com/')
```

```python
#clicking on explore
explore=driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]/details[1]/summary[1]")
explore.click()
```

```python
#clicking on trending
trending=driver.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a")
driver.get(trending.get_attribute('href'))
```

```python
#creating empty list
Repository_title=[]
Repository_description=[]
Contributors_count=[]
Language_used=[]
```

```python
title = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]')
for i in title:
    Repository_title.append(i.text)
Repository_title[:5]
```

```python
try:
    descriptions=driver.find_elements_by_xpath("//article[@class='Box-row']/p")
    for i in descriptions:
        Repository_description.append(i.text)
except NoSuchElementException:
    Repository_description.append('No details available')
Repository_description[:5]
```

```python
languages=driver.find_elements_by_xpath("//span[@class='d-inline-block ml-0 mr-3']")
for i in languages:
    try:
        Language_used.append(i.text)
    except NoSuchElementException :
        Language_used.append("---")
Language_used[:5]
```

```python
URLS=[]
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    URLS.append(i.get_attribute("href"))

for URL in URLS:
    
    try:
        driver.get(URL)
        count=driver.find_element_by_xpath("//ul[@class='pagehead-actions flex-shrink-0 d-none d-md-inline']/li[3]/a[2]").text
        Contributors_count.append(count)
    except NoSuchElementException :
        Contributors_count.append("---")
```

```python
Contributors_count[:5]
```

```python
# creating the dataframe from the scraped data
Repositories=pd.DataFrame({})
Repositories['Repository_title']=Repository_title[0:22]
Repositories['Repository_description']=Repository_description[0:22] 
Repositories['Contributors_count']=Contributors_count[0:22] 
Repositories['Language_used']=Language_used[0:22] 
Repositories
```
